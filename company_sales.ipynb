```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("company_sales_data.csv")
```

```python
x = df['month_number'] 
y = df['total_profit']

month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.plot(x, y)

plt.xticks(ticks=x, labels=month_labels) 

plt.title('Total Profits by Month')
plt.xlabel('Month') 
plt.ylabel('Total Profit')



plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x, df['bathingsoap'])
plt.xticks(ticks=x, labels=month_labels)
plt.title('Bathing Soap Sales')
plt.ylabel('Sales')

plt.subplot(2,1,2)
plt.plot(x, df['facewash'])
plt.xticks(ticks=x, labels=month_labels)
plt.title('Face Wash Sales')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```
